```python
#matplotlib
from matplotlib import pyplot as plt
from matplotlib import style

from numpy import*
a=array(
    [1,2,3])
b=array([3,4,6])
plt.plot(a,b,label="line",color="r",linewidth=9)    #linear plot
plt.title("population")
plt.xlabel("age")
plt.ylabel("weight")
plt.grid(True,color="c")
print(plt.style.available)
style.use("Solarize_Light2")
plt.legend()

plt.show()
```

```python
#bargraph
a=array([1,3,5,7,11,12])
b=array([2,4,6,8,9,13])
c=array([2,4,6,8,9,13])
d=array([1,4,7,11,22,22])
plt.bar(a,b,label="pop1",color="g")
plt.bar(c,d,label="pop1",color="k")
plt.legend()
plt.show()
```

```python
#histogram
a=array([1,2,3,4,5])
b=array([2,4,7,4,7])
plt.hist(a,histtype="bar",rwidth=7,label="popu")
plt.legend()
plt.show()
```

```python
plt.scatter(a,b,marker="*",color="b",label="popu")
plt.legend()
plt.show()

```
